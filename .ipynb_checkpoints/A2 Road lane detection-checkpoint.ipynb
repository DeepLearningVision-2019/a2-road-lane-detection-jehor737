## Helper Functions

Below are some helper functions to help get you started.

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img, lines

def weighted_img(img, initial_img, alpha=0.8, beta=1., gamma=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, alpha, img, beta, gamma)
```

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

```python
# TODO: Build your pipeline that will draw lane lines segments on the test_images
#Blur
kernel_size=9
#Canny thresholds
lower_threshold = 150
high_threshold = 250
#Hough Lines
rho = 5 
theta = np.pi/60
threshold = 1
min_line_len = 5
max_line_gap = 120
#Final result
images_with_lines = []
#Color detection
lower_hue_yellow = np.array([20,50,100], dtype=np.uint8) 
upper_hue_yellow = np.array([100,255,255], dtype=np.uint8)
lower_hue_white = np.array([20,200,0], dtype=np.uint8)
upper_hue_white = np.array([255,255,255], dtype=np.uint8)
images_color = []
images_canny = []
images_gaussian = []
images_masked = []

def process_image(ima):
    image_copy = ima[0] #copy of the image 
    image_title = ima[1] #title of the image
    hls = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HLS) #Convert to HLS space
    white = cv2.inRange(hls,lower_hue_white,upper_hue_white)
    yellow = cv2.inRange(hls,lower_hue_yellow,upper_hue_yellow)
    color_combined = cv2.bitwise_or(white,yellow)#combine mask
    image_color_detected = cv2.bitwise_and(image_copy,image_copy, mask=color_combined)
    images_color.append((image_color_detected,image_title))
    #for region of interest
    vertices = np.array([[(100,image_copy.shape[0]),(image_copy.shape[1]*0.45,image_copy.shape[0]*0.6),(image_copy.shape[1]*0.55,image_copy.shape[0]*0.6),(image_copy.shape[1],image_copy.shape[0])]], np.int32)
    #Apply Canny
    roadEdges = canny(grayscale(image_color_detected),lower_threshold, high_threshold)
    images_canny.append((roadEdges,image_title))
    #Apply Gaussian Blur to "highlight" the edges returned by Canny function
    gaussian_image = gaussian_blur(roadEdges,kernel_size)
    images_gaussian.append((gaussian_image,image_title))
    #Get the region of interest
    masked_image = region_of_interest(gaussian_image, vertices)
    images_masked.append((masked_image,image_title))
    #Get Hough Lines
    image, lines = hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap)
    #Join image with lines
    line_image = weighted_img(image, image_copy)
    return (line_image,image_title)

for ima in images:
    images_with_lines.append(process_image(ima))
```

## Improve the draw_lines() function

**At this point, you should have the Hough line segments drawn onto the road. Extend your code to define a line to run the full length of the visible lane based on the line segments you identified with the Hough Transform. Try to average and/or extrapolate the line segments you've detected to map out the full extent of the lane lines. The output should draw a single, solid line over the left lane line and a single, solid line over the right lane line. The lines should start from the bottom of the image and extend out to the top of the region of interest.**

```python
# TODO: Build your pipeline that will draw complete lane lines on the test_images
images_lines_full = []
def draw_lanes(ima):
    image_copy = ima[0] #copy of the image 
    image_original = image_copy
    image_title = ima[1] #title of the image
    hls = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HLS) #Convert to HLS space
    white = cv2.inRange(hls,lower_hue_white,upper_hue_white)
    yellow = cv2.inRange(hls,lower_hue_yellow,upper_hue_yellow)
    color_combined = cv2.bitwise_or(white,yellow)#combine mask
    image_color_detected = cv2.bitwise_and(image_copy,image_copy, mask=color_combined)
    #for region of interest
    vertices = np.array([[(100,image_copy.shape[0]),(image_copy.shape[1]*0.45,image_copy.shape[0]*0.6),(image_copy.shape[1]*0.55,image_copy.shape[0]*0.6),(image_copy.shape[1],image_copy.shape[0])]], np.int32)
    #Apply Canny
    roadEdges = canny(grayscale(image_color_detected),lower_threshold, high_threshold)
    #Apply Gaussian Blur to "highlight" the edges returned by Canny function
    gaussian_image = gaussian_blur(roadEdges,kernel_size)
    #Get the region of interest
    masked_image = region_of_interest(gaussian_image, vertices)
    #Get Hough Lines
    imag, lines = hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap)
    # Filter right and left lines
    left_line = []
    right_line = []
    for line in lines:
        if line[0][0] < 500:
            left_line.append(line[0])
        else:
            right_line.append(line[0])
    left_line = np.array(left_line)
    right_line = np.array(right_line)
    # Mean points of each line
    x = []
    y = []
    for x1, y1, x2, y2 in left_line:
        x += [x1, x2]
        y += [y1, y2]
    #Extrapolate 
    fit = np.polyfit(x, y, 1)
    slope, constant = fit
    m1 = np.array([[150, 150* slope + constant, 450, 450 * slope + constant]])
    x = []
    y = []
    for x1, y1, x2, y2 in right_line:
        x += [x1, x2]
        y += [y1, y2]
    #Extrapolate 
    fit = np.polyfit(x, y, 1)
    slope, constant = fit
    m2 = np.array([[500, 500* slope + constant, 950, 950 * slope + constant]])
    m1 = np.floor(m1).astype(int)
    m2 = np.floor(m2).astype(int)
    # Draw new lines
    lines = np.array([m1,m2])
    draw_lines(image_original, lines, thickness=10)
    return (image_original, image_title)

for ima in images:
    images_lines_full.append(draw_lanes(ima))

f, plots = plt.subplots((len(images_lines_full)+3-1)//3, 3, figsize=(20,10))
plots = [plot for sublist in plots for plot in sublist]

for image, plot in zip(images_lines_full, plots):
    plot.set_title(image[1])
    plot.imshow(image[0])
```
